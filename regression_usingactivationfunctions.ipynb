```python
model=Sequential()
model.add(Dense(7,activation='relu',input_dim=7))
model.add(Dense(7,activation='relu'))
model.add(Dense(1,activation='linear'))
```

The number of parameters in the neural network comes from the weights and biases in each layer.

For a Dense layer, the number of parameters is calculated as: (number of inputs * number of neurons) + number of neurons (for biases)

Looking at the model.summary() output:

dense_9 (Dense): (7 * 7) + 7 = 49 + 7 = 56 parameters
dense_10 (Dense): (7 * 7) + 7 = 49 + 7 = 56 parameters
dense_11 (Dense): (7 * 1) + 1 = 7 + 1 = 8 parameters
Adding the parameters from each layer: 56 + 56 + 8 = 120 total parameter

```python
model.summary()
```

compiling,giving loss function,optimizer

```python
model.compile(optimizer='Adam',loss='mean_squared_error')
```

Starting of training (fit of X_train,y_Train,giving no of epochs and dividing validation data from traing data)

```python
history=model.fit(x_train_scale,y_train,epochs=50,validation_split=0.2)
```

```python
y_prediction=model.predict(x_test_scale)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_prediction)
```

plotting loss (vs) validation loss

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'],label='train loss')
plt.plot(history.history['val_loss'],label='validation loss')
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','validation'],loc='upper left')
plt.show()
```
